• Create a list of product dictionaries (name, price, quantity)
• Calculate total inventory value (price × quantity)
• Find most expensive & cheapest product
• Display summary report: all items in dictionary, expensive, cheap

```python
products_dict =[
    { "name": "Laptop", "price": 1000, "stock": 50 },
    { "name": "Smartphone", "price": 500, "stock": 100 },
    { "name": "Tablet", "price": 300, "stock": 75 }
] 
for product in products_dict:
    total_value = product["price"] * product["stock"]
    
min_price = min([product["price"] for product in products_dict])
max_price = max([product["price"] for product in products_dict])
print(f"Minimum price: ${min_price}")
print(f"Maximum price: ${max_price}")
print(products_dict)
```
