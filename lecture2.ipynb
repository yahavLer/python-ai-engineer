## 🔨 Mini-Project 2: Sales Report Generator
### Build a script that processes sales data with loops:

- Create list of daily sales amounts
- Loop through and sum total revenue
- Find highest & lowest sales day
- Calculate average daily sales

```python
daily_sales = [100, 150, 200, 250, 300]
total_sales = sum(daily_sales)
average_sales = total_sales / len(daily_sales)
print(f"Total Sales: {total_sales}")
print(f"Average Sales: {average_sales}")
min_sales = daily_sales[0]
max_sales = daily_sales[0]
min_index = 0
max_index = 0
sum_sales = 0
for index, sale in enumerate(daily_sales):
    if sale < min_sales:
        min_sales = sale
        min_index = index
    if sale > max_sales:
        max_sales = sale
        max_index = index
    sum_sales += sale
print(f"Minimum Sale: {min_sales} at day {min_index}")
print(f"Maximum Sale: {max_sales} at day {max_index}")
print(f"Sum of Sales: {sum_sales}")
```

```python
list_days_sales = [[100, 150, 200], [250, 300, 350], [400, 450, 500]]
total_sales = 0
min_sale = list_days_sales[0][0]
max_sale = list_days_sales[0][0]
min_index = (0, 0)
max_index = (0, 0)
for index, day_sales in enumerate(list_days_sales):
    total_sales_day = 0
    for sale in day_sales:
        total_sales_day += sale
        if sale < min_sale:
            min_sale = sale
            min_index = (index, day_sales.index(sale))
        if sale > max_sale:
            max_sale = sale
            max_index = (index, day_sales.index(sale))
    print(f"Total Sales for Day {index}: {total_sales_day}")
    total_sales += total_sales_day

print(f"Total Sales: {total_sales}")
print(f"Minimum Sale: {min_sale} at index {min_index}")
print(f"Maximum Sale: {max_sale} at index {max_index}")
```

```python
def calculate_total_sales(sales_data) -> int:
    """Calculate the total sales from a list of daily sales.
    Args:
        sales_data (list of list of int): A list where each element is a list of sales for a day.
    Returns:
        int: The total sales across all days.
    """
    total_sales = 0
    for day_sales in sales_data:
        total_sales += sum(day_sales)
    return total_sales

total = calculate_total_sales(list_days_sales)

print(f"Total sales: {total}")
help(calculate_total_sales)
print(calculate_total_sales.__doc__)
```

### Mini-Project 3: Invoice Calculator
#### Build reusable functions for business calculations:

1. Function: calculate_subtotal(items_prices,
quantities)
2. Function: apply_tax(subtotal, tax_rate)
3. Function: final_invoice(subtotal, tax, discount=0)
4. Test with multiple orders, use different tax rates

```python
def calculate_subtotal(items_prices, quantities):
    """Calculate the subtotal for a list of items given their prices and quantities.
    Args:
        items_prices (list of float): A list of item prices.
        quantities (list of int): A list of quantities corresponding to each item.
    Returns:
        float: The subtotal for the items.
    """
    subtotal=0.0
    for index, price in enumerate(items_prices):
        quantity = quantities[index]
        subtotal += price * quantity
    return subtotal

prices = [10.0, 20.0, 30.0]
quantities = [1, 2, 3]
subtotal = calculate_subtotal(prices, quantities)
print(f"Subtotal: {subtotal}")
```

```python
def apply_tax(subtotal, tax_rate):
    """Apply tax to a subtotal to calculate the total amount.
    Args:
        subtotal (float): The subtotal amount before tax.
        tax_rate (float): The tax rate to apply. Defaults to 0.07 (7%).
    Returns:
        float: The total amount after applying tax.
    """
    total = subtotal * (1 + tax_rate)
    return total

total_amount_with_custom_tax = apply_tax(subtotal, 0.1) # Applying a custom tax rate of 10%
print(f"Total Amount after Custom Tax: {total_amount_with_custom_tax}")
```

```python
def final_invoice(subtotal, tax, discount=0):
    """Calculate the final invoice amount after applying tax and discount.
    Args:
        subtotal (float): The subtotal amount before tax and discount.
        tax (float): The tax rate to apply.
        discount (float): The discount rate to apply. Defaults to 0 (no discount).
    Returns:
        float: The final invoice amount after applying tax and discount.
    """
    total = subtotal * (1 + tax)
    final_amount = total - (total * discount)
    return final_amount

final_amount_with_discount = final_invoice(subtotal, 0.07, discount=0.05) # Applying a 5% discount
print(f"Final Invoice Amount after Discount: {final_amount_with_discount}")
```
