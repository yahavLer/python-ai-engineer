# Making HTTP Requests
Demonstrates GET and POST requests, error handling, and working with API responses.

```python
import requests
import json
```

```python
# Simple GET request
response = requests.get("https://api.github.com/users/github")
print(response.status_code)  # 200 = success
print(response.json())
```

```python
# GET request with headers and error handling
get_url = "https://jsonplaceholder.typicode.com/todos/1"
headers = {
    "Authorization": "Bearer YOUR_ACCESS_TOKEN",  # Replace with a real token if needed
    "User-Agent": "My-Python-App/1.0"
}

try:
    get_response = requests.get(get_url, headers=headers)
    get_response.raise_for_status()
    print("\n--- GET Request Success ---")
    print(f"Status Code: {get_response.status_code}")
    print(f"Title: {get_response.json().get('title')}")
except requests.exceptions.HTTPError as e:
    print(f"\nGET HTTP Error: {e.response.status_code}")
    print(f"Reason: {e}")
except requests.exceptions.RequestException as e:
    print(f"\nGET Request Error: {e}")
```

```python
# POST request example
BASE_URL = "https://jsonplaceholder.typicode.com/posts"
my_data = {
    "title": "My Unique Post Title",
    "body": "This is the body of the data I am posting.",
    "userId": 999
}

try:
    post_response = requests.post(BASE_URL, json=my_data)
    post_response.raise_for_status()
    posted_resource = post_response.json()   # This will raise an error since the response is JSON, not XML
    print("--- POST Request SUCCESS ---")
    print(f"Status Code: {post_response.status_code}")
    print(f"Your Posted Data (from POST response): {posted_resource}")
except requests.exceptions.RequestException as e:
    print(f"POST Request Error: {e}")
```

# Handling API Responses
Reusable function to fetch API data, handle errors, and save the response to a file.

```python
def fetch_api_data(url, headers=None):
    try:
        response = requests.get(url, headers=headers, timeout=5)
        response.raise_for_status()  # Raise on 4xx/5xx
        data = response.json()
        # Save to file
        with open("response.json", "w") as f:
            json.dump(data, f, indent=2)
        return data
    except requests.exceptions.Timeout:
        print("Request timed out")
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e.response.status_code}")
    except json.JSONDecodeError:
        print("Invalid JSON response")
    except Exception as e:
        print(f"Unexpected error: {e}")
```

```python
# Example usage of fetch_api_data
url = "https://api.github.com/users/github"
data = fetch_api_data(url)
pretty_data = json.dumps(data, indent=4)
print(pretty_data)
```

# Reading Files Effectively
Examples for reading text, CSV, and JSON files in Python.

```python
# Read entire file
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Read line by line
with open("data.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Read CSV files
import csv
with open("products.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row["name"], row["price"])
```

```python
# Read JSON files
with open("config.json", "r") as file:
    config = json.load(file)
    print(config["api_key"])
```
