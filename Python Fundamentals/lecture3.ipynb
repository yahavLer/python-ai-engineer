### Mini Project 1- Weather Calculator
1. Use random to simulate 24 daily temperatures
2. Use math to calculate average, max, min
3. Use datetime to record date/time
4. Convert Celsius to Fahrenheit using math formula

```python
temperature = []
for i in range(25):
    temp= random.randint(1,100)
    temperature.append(temp)

avg= sum(temperature) / len(temperature)
min_temp = min(temperature)
max_temp = max(temperature)
cure_time = datetime.now()

temperature_fahrenheit=[]
for temp in temperature:
    temperature_fahrenheit.append((temp * 9/5) + 32)

print("===== Weather Calculator =====")
print("Date & Time:", cure_time)

print("\nHourly Temperatures (°C):")
print(temperature)

print("\nHourly Temperatures (°F):")
print(temperature_fahrenheit)

print("\nWeather Summary:")
print("Average Temperature:", avg, "°C")
print("Maximum Temperature:", max_temp, "°C")
print("Minimum Temperature:", min_temp, "°C")
```

```python
class Customer:
    def __init__(self, id, name, email, balance=0):
        self.id = id
        self.name = name
        self.email = email
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        return self.balance
    def withdraw(self, amount):
        if amount > self.balance:
            raise ValueError("Insufficient funds")
        self.balance -= amount
        return self.balance
    

customer = Customer(1, "Alice", "alice@company.com", 1000)
customer.deposit(500)
customer.withdraw(200)
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def apply_discount(self, rate):
        return self.price * (1 - rate)

    def info(self):
        return f"{self.name}: ${self.price}"


# Create objects
laptop = Product("Laptop", 1000)
mouse = Product("Mouse", 50)

print(laptop.info())
print(f"Discounted: ${laptop.apply_discount(0.2)}")
print(type(laptop))
print(type(mouse))
```

### Inheritance Example: Cat

```python
class Cat(Animal): # Cat inherits from Animal
    def __init__(self, name, age, mustach_length):
        super().__init__(name, age) ##call father constractor
        self.mustach_length = mustach_length

    def sound(self):
        print(f"{self.name} says: Meuw!")

    def scratch(self):
        print(f"{self.name} is scrstching")


# Usage
my_cat = Cat("Whiskers", 3, 15)
my_cat.eat() # Inherited from Animal
my_cat.sleep() # Inherited from Animal
my_cat.sound() # New method in Cat
my_cat.scratch() # New method in Cat
```

```python
class Dog(Animal): # Dog inherits from Animal
    def sound(self):
        print(f"{self.name} says: Woof!")
    def fetch(self):
        print(f"{self.name} is fetching the ball")


my_dog = Dog("Rex", 5)
my_dog.eat() # Inherited from Animal
my_dog.sleep() # Inherited from Animal
my_dog.sound() # New method in Dog
my_dog.fetch() # New method in Dog
```

### Mini-Project 2, part 2: inheritance

Dog has all Animal methods + its own: bark() and fetch()
1. create dog and cat the inherit from animal
2. create sound method for both
3. create scratch method for cat
4. create fetch method for dog

```python
# Parent Class (Animal) - Based on Slide 25 & 27
class Animal:
    """Base class for animals."""
    def __init__(self, name: str, age: int) -> None:
        self.name = name
        self.age = age

    def eat(self) -> str:
        return f"{self.name} is eating."

    def sleep(self) -> str:
        return f"{self.name} is sleeping."

# Child Class 1 (Cat) - Inherits from Animal - Based on Slide 28
class Cat(Animal):
    """Cat adds `mustache_length` and cat-specific behavior."""
    def __init__(self, name: str, age: int, mustache_length: float) -> None:
        super().__init__(name, age)
        self.mustache_length = mustache_length

    def sound(self) -> str:
        return f"{self.name} says: Meow!"

    def scratch(self) -> str:
        return f"{self.name} is scratching."

# Child Class 2 (Dog) - Inherits from Animal - Based on Slide 30
class Dog(Animal):
    """Dog explicitly calls super().__init__ even if no extra fields."""
    def __init__(self, name: str, age: int) -> None:
        super().__init__(name, age)

    def sound(self) -> str:
        return f"{self.name} says: Woof!"
    
    def fetch(self) -> str:
        return f"{self.name} is fetching the ball."

def main() -> None:
    # create animals and demonstrate polymorphism

    cat1 = Cat("Whiskers", 3, 5.0)
    cat2 = Cat("Luna", 1, 2.5)
    dog1 = Dog("Rex", 5)
    dog2 = Dog("Buddy", 2)

    print("\n--- Print single cat and dog (uses __str__) ---")
    print(cat1.eat())
    print(cat1.sleep())
    print(cat1.sound())
    print(cat1.scratch())
    print(f"Mustache length: {cat1.mustache_length} cm")
    print(Cat.__mro__) #mathod resulotion order- how python know who call
    print("-----")
    print(dog1.eat())
    print(dog1.sleep())
    print(dog1.sound())
    print(dog1.fetch())

    animals: List[Animal] = [cat1, cat2, dog1, dog2]

    print("\n--- Polymorphism demo: iterate animals and call methods ---")
    for a in animals:
        print(a.eat())
        print(a.sleep())
        print(a.sound())
        if isinstance(a, Cat):
            print(a.scratch())
        if isinstance(a, Dog):
            print(a.fetch())


if __name__ == "__main__":
    main()
```
