# חלק 2 — מחלקות ו־OOP

במקום לשמור מוצר כמילון בלבד, ניצור מחלקה בשם `Product`.

מחלקה מאפשרת לשמור יחד:

- מידע: שם, מחיר, כמות.
- פעולות: חישוב ערך מלאי, החלת הנחה, יצירת תיאור.

</div>

````python
class Product:
    def __init__(self, name, price, quantity):
        # TODO: save name, price, and quantity as attributes
        self.name = name
        self.price = price
        self.quantity = quantity

    def inventory_value(self):
        # TODO: return price multiplied by quantity
        return self.price * self.quantity

    def apply_discount(self, discount):
        # TODO: return price after discount
        return self.price * (1 - discount)

    def info(self):
        # TODO: return a dictionary with product data
        return {
            "name": self.name,
            "price": self.price,
            "quantity": self.quantity,
            "inventory_value": self.inventory_value()
        }


laptop = Product("Laptop", 3500, 4)
mouse = Product("Mouse", 80, 20)

print(laptop.inventory_value())
print(mouse.apply_discount(0.1))
print(laptop.info())
````

# חלק 3 — דיבוג וטיפול בשגיאות

כעת נשפר את המחלקה כך שלא תאפשר מחיר שלילי או כמות שלילית.

</div>

````python
class SafeProduct:
    def __init__(self, name, price, quantity):
        # TODO: raise ValueError if price is not positive
        if price <= 0:
            raise ValueError("Price must be positive")

        # TODO: raise ValueError if quantity is negative
        if quantity < 0:
            raise ValueError("Quantity cannot be negative")

        self.name = name
        self.price = price
        self.quantity = quantity

    def inventory_value(self):
        return self.price * self.quantity
    
    def info(self):
        return {
            "name": self.name,
            "price": self.price,
            "quantity": self.quantity,
            "inventory_value": self.inventory_value()
        }


try:
    product = SafeProduct("Broken Product", -10, 3)
    print(product.inventory_value())
except ValueError as error:
    print(f"Invalid product: {error}")

try:
    product = SafeProduct("Broken Product", 10, -3)
    print(product.inventory_value())
except ValueError as error:
    print(f"Invalid product: {error}")

try:
    product = SafeProduct("Broken Product", 10, 3)
    print(product.inventory_value())
except ValueError as error:
    print(f"Invalid product: {error}")
````

# חלק 4 — שמירה וטעינה של JSON

JSON הוא פורמט נפוץ מאוד לשמירת מידע ולעבודה עם API.

### משימה 4

שמרו את דוח המלאי לקובץ בשם `inventory_report.json`, ואז קראו אותו חזרה.

</div>

````python
import json

products = [
    SafeProduct("Laptop", 3500, 4),
    SafeProduct("Mouse", 80, 20),
    SafeProduct("Keyboard", 250, 10)
]

report = {
    "created_at": str(datetime.now()),
    "products": [],
    "total_value": 0
}

for product in products:
    # TODO: append product data to report["products"]
    report["products"].append(product.info())

    # TODO: add product inventory value to total_value
    report["total_value"] += product.inventory_value()

# TODO: save report to JSON file
with open("inventory_report.json", "w") as file:
    json.dump(report, file, indent=4)

# TODO: load report from JSON file
with open("inventory_report.json", "r") as file:
    loaded_report = json.load(file)

print(loaded_report)
````

# חלק 5 — קריאה ל־API עם requests

נשתמש ב־JSONPlaceholder, שהוא API חינמי לתרגול.

</div>

````python
import requests


def fetch_todo(todo_id):
    url = f"https://jsonplaceholder.typicode.com/todos/{todo_id}"

    try:
        # TODO: send GET request with timeout=5
        response = requests.get(url, timeout=5)

        # TODO: raise error for bad HTTP status
        response.raise_for_status()

        # TODO: convert response to JSON
        data = response.json()
        return data

    except requests.exceptions.Timeout:
        print("Request timed out")

    except requests.exceptions.RequestException as error:
        print(f"Request failed: {error}")

    return None


todo = fetch_todo(1)

if todo is not None:
    print(todo)
    print(todo["title"])
else:
    print("No data returned")
````

# חלק 6 — פרויקט מסכם קטן

כעת נשלב הכול: מחלקה, פונקציות, JSON, טיפול בשגיאות, ודוח סופי.

### משימה מסכמת

השלימו את הקוד כך שייצור דוח מלאי מלא.

</div>

````python
class InventoryManager:
    def __init__(self):
        # TODO: initialize empty products list
        self.products = []

    def add_product(self, product):
        # TODO: add product to list
        self.products.append(product)

    def total_value(self):
        total = 0
        for product in self.products:
            # TODO: add inventory value
            total += product.inventory_value()
        return total

    def to_dict(self):
        return {
            "created_at": str(datetime.now()),
            "number_of_products": len(self.products),
            "total_value": self.total_value(),
            "products": [product.info() for product in self.products]
        }


manager = InventoryManager()
manager.add_product(SafeProduct("Laptop", 3500, 4))
manager.add_product(SafeProduct("Mouse", 80, 20))
manager.add_product(SafeProduct("Keyboard", 250, 10))

final_report = manager.to_dict()

with open("final_inventory_report.json", "w") as file:
    json.dump(final_report, file, indent=4)

print(final_report)
````

````python
class WeatherReport:
    def __init__(self, temperatures):
        self.created_at = str(datetime.now())
        self.temperatures = temperatures
        self.max_temp = max(temperatures)
        self.min_temp = min(temperatures)
        self.avg_temp = sum(temperatures) / len(temperatures)

    def to_dict(self):
        return {
            "created_at": self.created_at,
            "temperatures": self.temperatures,
            "max_temp": self.max_temp,
            "min_temp": self.min_temp,
            "avg_temp": self.avg_temp
        }
    
weather_report = WeatherReport(temperatures)
with open("weather_report.json", "w") as file:
    json.dump(weather_report.to_dict(), file, indent=4)

````
